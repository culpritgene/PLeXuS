```python
from scipy import optimize
def f(x,a,b,c):
    return a * np.exp(b * x) + c 
```

```python
coef, cov =optimize.curve_fit(f, xdata=pa.graph.vs()['betwenness_z'], ydata=pa.graph.vs()['Ref_number'])
```

```python
f2 = lambda x: f(x,*coef) 
a = (np.log(np.array(pa.graph.vs()['Ref_number']))-np.log(np.array(list(map(f2, pa.graph.vs()['betwenness_z'])))))
b = a/a.std()
```
